```python
RepaymentsDF = pd.read_csv("D:\AnalytixLabs\Case Studies\Python Case Study\Case Study 2 - Credit Card Case Study\credit-risk-and-spending-analysis\Repayment.csv", index_col= "SL No:")
```

```python
spendDF = pd.read_csv("D:\AnalytixLabs\Case Studies\Python Case Study\Case Study 2 - Credit Card Case Study\credit-risk-and-spending-analysis\spend.csv", index_col = "Sl No:")
```

```python
Customer_acqDF[Customer_acqDF.Age < 18] = Customer_acqDF.Age.mean()
```

```python
temp_df = pd.merge(left= spendDF, right = Customer_acqDF, left_on= 'Customer', right_on='Customer',how = 'inner')

final_df = pd.merge(temp_df, RepaymentsDF, how= 'inner', on= 'Customer', suffixes=['_spend', '_Repayments' ])
```

```python
# conversion 
final_df = final_df.convert_dtypes()
final_df.Month_spend = pd.to_datetime(final_df.Month_spend)

final_df.Month_Repayments = pd.to_datetime(final_df.Month_Repayments)

final_df['Month'] = final_df.Month_spend.dt.month

final_df['Month_Name'] = final_df.Month_spend.dt.month_name()
```

```python
final_df['Amount_spend'] = final_df['Amount_spend'].where(
    final_df['Amount_spend'] <= final_df['Limit'],
    final_df['Limit'] / 2
)
```

```python
final_df
```

```python
final_df['Amount_Repayments'] = final_df['Amount_Repayments'].where(final_df['Amount_Repayments'] <= final_df['Limit'], final_df['Limit'])
```
